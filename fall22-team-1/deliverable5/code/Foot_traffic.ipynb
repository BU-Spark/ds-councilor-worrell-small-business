```python
import pandas as pd
```

```python
df_FieldsCorner = pd.read_csv("Fields Corner.csv")
df_FourCorners = pd.read_csv("Four Corners.csv")
df_GreaterAshmont = pd.read_csv("Greater Ashmont.csv")
df_GreaterGroveHall = pd.read_csv("Greater Grove Hall.csv")
df_Brighton = pd.read_csv("Brighton.csv")
df_Chinatown = pd.read_csv("Chinatown.csv")
df_EastBoston = pd.read_csv("East Boston.csv")
df_EglestonSquare = pd.read_csv("Egleston Square.csv")
df_HydePark = pd.read_csv("Hyde Park.csv")
df_JPCentreSouth = pd.read_csv("JP Centre South.csv")
df_MattapanSquare = pd.read_csv("Mattapan Square.csv")
df_MissionHill = pd.read_csv("Mission Hill.csv")
df_RoslindaleVillage = pd.read_csv("Roslindale Village.csv")
df_Roxbury = pd.read_csv("Roxbury.csv")
df_ThreeSquares = pd.read_csv("Three Squares.csv")
df_UphamsCorner = pd.read_csv("Uphams Corner.csv")
df_WashingtonGateway = pd.read_csv("Washington Gateway.csv")
df_WestRoxbury = pd.read_csv("West Roxbury.csv")
```

```python
df = df_FieldsCorner.copy()
df["total"] = (df_JPCentreSouth["Percent Change"]+df_MattapanSquare["Percent Change"]+df_RoslindaleVillage["Percent Change"])/4
month_footTraffic = pd.DataFrame({"Month":"","Percent Change":""},index=[])
for i in range(int(len(df)/4)):
    mean_per_month = df[i*4:i*4+4]["total"].mean()
    time = df["Week Ending"][i*4]
    time_split = time.split("/")
    month_time = time_split[0]+"/"+time_split[2][-2:]
    month_footTraffic.loc[i]=[month_time,mean_per_month]
month_footTraffic
```

```python
import matplotlib.pyplot as plt
mean=df["total"].mean()
plt.figure(figsize = (15,10))
plt.plot( month_footTraffic["Month"], month_footTraffic["Percent Change"], label='Percentage Change', linewidth=1, color='blue', marker='o', markerfacecolor='red', markersize=1)
plt.axhline(y=mean, color = 'y', label="Mean")
plt.axhline(y=0, color = 'g', label = "Percentage Change = 0")
plt.legend()
plt.xlabel("Month",size=20)
plt.ylabel("Percentage Change",size=20)
plt.title("Foot Traffic Boston",size=20)
plt.show()
```

```python
from sklearn.cluster import KMeans

dis = []
for i in range(1,10):
    kmean = KMeans(n_clusters=i, random_state=0, n_init = 50, max_iter = 500)
    kmean.fit(df[['total']])
    dis.append(kmean.inertia_)
plt.plot(range(1,10), dis)
plt.xlabel('k')
plt.show()
```

```python
kmean = KMeans(n_clusters=3, random_state=0, n_init = 50, max_iter = 500)
kmean.fit(df[['total']])
kmean.labels_
```
