```python
#import the license datasets from the datasets/licenses folder
import pandas as pd
import matplotlib.pyplot as plt

ote_licenses = pd.read_csv('../datasets/licenses/one_time_entertainment_licenses.csv')

spe_licenses = pd.read_csv('../datasets/licenses/special_event_licenses.csv')

ent_licenses = pd.read_csv('../datasets/licenses/entertainment_licenses.csv')

spe_licenses = spe_licenses[spe_licenses['City'] == 'Boston']
ent_licenses = ent_licenses[ent_licenses['city'] == 'Boston']
ote_licenses = ote_licenses[ote_licenses['city'] == 'Boston']

# district4Zips = [ '01747', '01748', '01757', '02019', '02030', '02035', '02038', '02048', '02052', '02053', '02054', '02056', 
# '02062', '02067', '02071', '02072', '02081', '02093', '02115', '02130', '02135', '02215', '02346', '02347', '02356', '02357', '02375', 
# '02445', '02446', '02457', '02458', '02459', '02460', '02461', '02462', '02464', '02465', '02466', '02467', '02468', '02472', '02481', 
# '02482', '02492', '02494', '02702', '02703', '02715', '02717', '02718', '02720', '02721', '02723', '02725', '02726', '02760', '02762', 
# '02763', '02764', '02766', '02767', '02769', '02771', '02777', '02779', '02780', '02790', '02861', '02914' ]

district4Zips = ['02126', '02124' , '02122', '02121', '02130', '02132', '02467', '02119', '02118', '02115']


#place the bar graphs side by side, include a legend, and label the x and y axes
fig, ax = plt.subplots(1, 3, figsize=(25, 8))
#set a title for the figure
fig.suptitle('Top Zip Codes for License Types in all of Boston', fontsize=16)
spe_licenses['Zip Code'].value_counts().head(10).plot(kind='bar',color = 'red', ax=ax[0])
ax[0].set_title('Special Event Licenses')
ax[0].set_xlabel('Zip Code')
ax[0].set_ylabel('Number of Licenses')
ax[0].set_xticklabels(spe_licenses['Zip Code'].value_counts().head(10).index, rotation=45)
ent_licenses['zip'].value_counts().head(10).plot(kind='bar',color = 'green', ax=ax[1])
ax[1].set_title('Entertainment Licenses')
ax[1].set_xlabel('Zip Code')
ax[1].set_ylabel('Number of Licenses')
ax[1].set_xticklabels(ent_licenses['zip'].value_counts().head(10).index, rotation=45)
ote_licenses['zip'].value_counts().head(10).plot(kind='bar', ax=ax[2])
ax[2].set_title('One Time Entertainment Licenses')
ax[2].set_xlabel('Zip Code')
ax[2].set_ylabel('Number of Licenses')
ax[2].set_xticklabels(ote_licenses['zip'].value_counts().head(10).index, rotation=45)
plt.show()









```

```python
#go through the zip code column in the entertainment licenses dataset and add a 0 to the beginning of each zip code
ent_licenses['zip'] = ent_licenses['zip'].apply(lambda x: '0' + str(x) if len(str(x)) == 4 else str(x))

#go through the zip code column in the one time entertainment licenses dataset and add a 0 to the beginning of each zip code while also deleting the last 2 digits
ote_licenses['zip'] = ote_licenses['zip'].apply(lambda x: '0' + str(x)[:-2] if len(str(x)) == 6 else str(x))


#remove all the zip codes that are not in district 4
spe_licenses = spe_licenses[spe_licenses['Zip Code'].isin(district4Zips)]
ent_licenses = ent_licenses[ent_licenses['zip'].isin(district4Zips)]
ote_licenses = ote_licenses[ote_licenses['zip'].isin(district4Zips)]

#place the bar graphs side by side, include a legend, and label the x and y axes
fig, ax = plt.subplots(1, 3, figsize=(25, 8))
#set a title for the figure
fig.suptitle('Top Zip Codes for License Types in District 4', fontsize=16)
spe_licenses['Zip Code'].value_counts().head(10).plot(kind='bar',color = 'red', ax=ax[0])
ax[0].set_title('Special Event Licenses')
ax[0].set_xlabel('Zip Code')
ax[0].set_ylabel('Number of Licenses')
ax[0].set_xticklabels(spe_licenses['Zip Code'].value_counts().head(10).index, rotation=45)
ent_licenses['zip'].value_counts().head(10).plot(kind='bar',color = 'green', ax=ax[1])
ax[1].set_title('Entertainment Licenses')
ax[1].set_xlabel('Zip Code')
ax[1].set_ylabel('Number of Licenses')
ax[1].set_xticklabels(ent_licenses['zip'].value_counts().head(10).index, rotation=45)
ote_licenses['zip'].value_counts().head(10).plot(kind='bar', ax=ax[2])
ax[2].set_title('One Time Entertainment Licenses')
ax[2].set_xlabel('Zip Code')
ax[2].set_ylabel('Number of Licenses')
ax[2].set_xticklabels(ote_licenses['zip'].value_counts().head(10).index, rotation=45)
plt.show()
```
