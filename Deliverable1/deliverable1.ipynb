```python
import pandas as pd
import folium 

def read_csv(filepath):
    df = pd.read_csv(filepath)
    return df

df = read_csv("../datasets/MainStreets_Business_List.csv")

#plot every business on a map
def plot_businesses(df):
    map = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], zoom_start=13)
    for i in range(len(df)):
        folium.Marker([df.iloc[i]['latitude'], df.iloc[i]['longitude']]).add_to(map)
    return map

plot_businesses(df)
```
