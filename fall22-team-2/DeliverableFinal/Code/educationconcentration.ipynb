The client wanted to know what areas/neighborhoods have the highest concentration of "high-value" workers (workers with higher education degrees). 
We created graphs to compare the employee education profile by neighborhoods to see which neighborhoods have the highest concentration. 
We then created graphs for employee income levels by neighborhood to see if there is a correlation between the education levels of employees and income.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("d4_neighborhoods_employment.csv")
df = df.iloc[[35, 36, 37, 38, 39]]




df.plot(x="Work Area Profile Report", y=["Mattapan", "Roslindale", "HydePark", "FieldsCorner"], kind="barh",legend=True, width=0.8,figsize=(10,10))
plt.xlabel("Number of Employees", fontsize=15)
plt.ylabel("Education Profile", fontsize=15)
plt.title("District 4: Employee Education Profile by Neighborhood",fontsize=15)
# plt.bar_label(container=, label_type='center', padding=0)
plt.show()


```

```python
df = pd.read_csv("otherDistrict_employment.csv")
df = df.iloc[[35, 36, 37, 38, 39]]

df.plot(x="Work Area Profile Report", y=["Allston", "Chinatown", "UphamsCorner", "WestRoxbury","ThreeSquares"], kind="barh",legend=True, width=0.8,figsize=(10,10), color=["pink", "grey", 'cyan', 'olive', 'purple'])
plt.xlabel("Number of Employees", fontsize=15)
plt.ylabel("Education Profile", fontsize=15)
plt.title("Other Districts: Employee Education Profile by Neighborhood",fontsize=15)
plt.show()
```

```python
df = pd.read_csv("otherDistrict_employment.csv")
df = df.iloc[[35, 36, 37, 38, 39]]

df.plot(x="Work Area Profile Report", y=["GreaterAshmont", "GreaterGroveHall", "JPCentreSouth", "Egleston", "FourCorners", "BowdoinGeneva"], kind="barh",legend=True, width=0.8,figsize=(10,10),color=['gold', 'salmon', 'palegreen', 'teal', 'navy', 'plum'])
plt.xlabel("Number of Employees", fontsize=15)
plt.ylabel("Education Profile", fontsize=15)
plt.title("Other Districts: Employee Education Profile by Neighborhood",fontsize=15)
plt.show()
```
