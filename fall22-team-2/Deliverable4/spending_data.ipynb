Goal: Observe weekly spending and influences from Covid
- pre-covid : Jan 5 2020 until March 8 2020
- during covid : March 15 2020 until November 15 2020

` Using data from Mattapan neighborhood ` 

```python
mattapanSpending = pd.read_csv("Spending Data\Mattapan Weekly.csv")
mattapanSpending.rename(columns={'Row Labels': 'Date'}, inplace=True)
```

```python
# remove the %
mattapanSpending["Change in In-Person Mastercard Spending"] = mattapanSpending["Change in In-Person Mastercard Spending"].str[:-1]
#convert to int
mattapanSpending["Change in In-Person Mastercard Spending"] = mattapanSpending["Change in In-Person Mastercard Spending"].astype('int')
```

```python
mattapanSpending
```

```python
mattapanSpending.plot(x = "Date", y = "Change in In-Person Mastercard Spending", kind='line', title="Mattapan Weekly Spending Data")
```
