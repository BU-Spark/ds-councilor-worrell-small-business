# Estimated_Employment #

We look at the `estimated_employment` column of the businesses dataset for multiple districts. 

```python
import pandas as pd
import matplotlib.pyplot as plt

business_df = pd.read_csv('district4_businesses.csv')
business_df['estimated_employment'].value_counts().plot(kind='barh',legend=True, width=0.8,figsize=(10,10))

plt.xticks(fontsize=15,color='white')
plt.yticks(fontsize=15, color='white')
plt.ylim(-1,32)
plt.xlim(0, 75)
plt.xlabel("Number of Businesses", color='white', fontsize=15)
plt.ylabel("Estimated_Employment", color='white',fontsize=15)
plt.title("District 4: Estimated Employment per Small Business",color='white',fontsize=15)
plt.show()
```
